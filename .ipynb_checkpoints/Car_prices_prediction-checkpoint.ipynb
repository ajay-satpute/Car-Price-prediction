## Problem Statement :
A Chinese automobile company Geely Auto aspires to enter the US market by setting up their manufacturing unit there and producing cars locally to give competition to their US and European counterparts. 

Specifically, they want to understand the factors affecting the pricing of cars in the American market, since those may be very different from the Chinese market. The company wants to know:
   - Which variables are significant in predicting the price of a car
   - How well those variables describe the price of a car
Based on various market surveys, the consulting firm has gathered a large dataset of different types of cars across the Americal market.

## Business Goal :

We are required to model the price of cars with the available independent variables. It will be used by the management to understand how exactly the prices vary with the independent variables. They can accordingly manipulate the design of the cars, the business strategy etc. to meet certain price levels. Further, the model will be a good way for management to understand the pricing dynamics of a new market.

## Step 3: Data Preparation

```python
# Identifying numerical columns
num_cols = car_price._get_numeric_data().columns
num_cols
```

```python
# Identifying categorical columns
cat_cols = car_price.select_dtypes(include=['O']).columns.values
cat_cols
```

<font color='red'>**From data visualisation it is clear that `car_ID`, `symboling`, `carheight`,`stroke`, `compressionratio`, `peakrpm` columns are irrelevant to our model.**</font>

```python
# Deleting irrelevant columns
car_price.drop(['car_ID','symboling','carheight', 'stroke', 'compressionratio', 'peakrpm'], axis = 1, inplace = True)
```

There is a variable named CarName which is comprised of two parts - the first word is the name of 'car company' and the second is the 'car model'. For example, chevrolet impala has 'chevrolet' as the car company name and 'impala' as the car model name. We need to consider only company name as the independent variable for model building. 

```python
# Retaining only company name from CarName column

car_price['CarName'] = car_price['CarName'].apply(lambda x: x.split(' ')[0])

# Lets see unique values in column 'CarName'

print('List of unique values in column CarName are :\n\n', car_price['CarName'].unique())
print('\nTotal no. of unique values in column CarName are :',car_price['CarName'].nunique())
```

```python
# Converting all names to lower case
car_price['CarName'] = car_price['CarName'].apply(lambda x: x.lower())

# Some company names are wrongly spelled
# Replacing 'maxda' to 'mazda', 'porcshce' to 'porsche','toyouta' to 'toyota' and 'vokswagen', 'vw' to 'volkswagen'

car_price['CarName'] = car_price['CarName'].replace({'maxda':'mazda'},regex=True)
car_price['CarName'] = car_price['CarName'].replace({'porcshce':'porsche'},regex=True)
car_price['CarName'] = car_price['CarName'].replace({'toyouta':'toyota'},regex=True)
car_price['CarName'] = car_price['CarName'].replace({'vokswagen':'volkswagen','vw':'volkswagen'},regex=True)

print('List of unique values in column CarName are :\n\n', car_price['CarName'].unique())
print('\nTotal no. of unique values in column CarName are :',car_price['CarName'].nunique())
```

### Converting `categorical columns` into `numerical columns`

```python
# Converting 'gas' to 0 and 'diesel' to 1 from 'fueltype' column
car_price['fueltype'] = car_price['fueltype'].replace({'gas':0,'diesel':1},regex=True)

# Converting 'std' to 0 and 'diesel' to 1 from 'aspiration' column
car_price['aspiration'] = car_price['aspiration'].replace({'std':0,'turbo':1},regex=True)

# Converting 'two' to 0 and 'four' to 1 from 'doornumber' column
car_price['doornumber'] = car_price['doornumber'].replace({'two':0,'four':1},regex=True)

# Converting 'front' to 0 and 'rear' to 1 from 'enginelocation' column
car_price['enginelocation'] = car_price['enginelocation'].replace({'front':0,'rear':1},regex=True)
```

### Dummy Variables

```python
# Get the dummy variables for the feature 'CarName' and store it in a new variable - 'company_name'
# and drop the first column from company_name df using 'drop_first = True'
company_name = pd.get_dummies(car_price['CarName'], drop_first = True)

# Add the results to the original car_price dataframe
car_price = pd.concat([car_price, company_name], axis = 1)

# Drop 'CarName' as we have created the dummies for it
car_price.drop(['CarName'], axis = 1, inplace = True)

# Using the same process to get the dummy variables for the features 'carbody', 'drivewheel', 'enginetype', 'fuelsystem' 
# and 'cylindernumber'.

Car_Body = pd.get_dummies(car_price['carbody'], drop_first = True)
car_price = pd.concat([car_price, Car_Body], axis = 1)
car_price.drop(['carbody'], axis = 1, inplace = True)

Drive_Wheel = pd.get_dummies(car_price['drivewheel'], drop_first = True)
car_price = pd.concat([car_price, Drive_Wheel], axis = 1)
car_price.drop(['drivewheel'], axis = 1, inplace = True)

Engine_Type = pd.get_dummies(car_price['enginetype'], drop_first = True)
car_price = pd.concat([car_price, Engine_Type], axis = 1)
car_price.drop(['enginetype'], axis = 1, inplace = True)

Fuel_System = pd.get_dummies(car_price['fuelsystem'], drop_first = True)
car_price = pd.concat([car_price, Fuel_System], axis = 1)
car_price.drop(['fuelsystem'], axis = 1, inplace = True)

Cylinder_Number = pd.get_dummies(car_price['cylindernumber'], drop_first = True)
car_price = pd.concat([car_price, Cylinder_Number], axis = 1)
car_price.drop(['cylindernumber'], axis = 1, inplace = True)

car_price.head()
```

## Step 4: Splitting the Data into Training and Testing Sets


```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
cp_train, cp_test = train_test_split(car_price, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
cp_train.describe()
```

### Rescaling the Features 

We will use MinMax scaling.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['wheelbase', 'carlength', 'carwidth', 'curbweight','enginesize', 'boreratio',
            'horsepower', 'citympg', 'highwaympg', 'price']

cp_train[num_vars] = scaler.fit_transform(cp_train[num_vars])
```

```python
cp_train.describe()
```

```python
cp_train.shape
```

### Dividing into X and Y sets for the model building

```python
y_train = cp_train.pop('price')
X_train = cp_train
```

## Step 5: Building a linear model


### RFE
Recursive feature elimination

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

# Running RFE with the output number of the variable equal to 16
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 16)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

### Building model using statsmodel, for the detailed statistics

```python
# Creating X_train dataframe with RFE selected variables
X_train = X_train[col]
```

```python
# Creating a function for building model

import statsmodels.api as sm  

def lr_model(b,a):
    a_const = sm.add_constant(a)
    lm = sm.OLS(b,a_const).fit()   # Running the linear model
    print(lm.summary())

# Deifining VIF funcion to calculate the VIFs for the new model

from statsmodels.stats.outliers_influence import variance_inflation_factor

def VIF(X):
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return(vif)
```

```python
#Creating a model

lr_model(y_train,X_train)
VIF(X_train)
```

<font color='red'>**Now from here onwards we are going to update the model by dropping variables which have p-value > 0.05 (Significance level) and vif > 5.**</font>

```python
# Dropping the variable 'twelve' and updating the model

X_train_new = X_train.drop(["twelve"], axis = 1)
#Creating a model

lr_model(y_train,X_train_new)
VIF(X_train_new)
```

```python
# Dropping the variable 'highwaympg' and updating the model

X_train_new = X_train_new.drop(["highwaympg"], axis = 1)
#Creating a model

lr_model(y_train,X_train_new)
VIF(X_train_new)
```

```python
# Dropping the variable 'dohcv' and updating the model

X_train_new = X_train_new.drop(["dohcv"], axis = 1)

# Creating a model
lr_model(y_train,X_train_new)
VIF(X_train_new)
```

```python
# Dropping the variable 'carwidth' and updating the model

X_train_new = X_train_new.drop(["carwidth"], axis = 1)

# Creating a model
lr_model(y_train,X_train_new)
VIF(X_train_new)
```

```python
# Dropping the variable 'three' and updating the model

X_train_new = X_train_new.drop(["three"], axis = 1)

# Creating a model
lr_model(y_train,X_train_new)
VIF(X_train_new)
```

```python
# Dropping the variable 'hardtop' and updating the model

X_train_new = X_train_new.drop(["hardtop"], axis = 1)

# Creating a model
lr_model(y_train,X_train_new)
VIF(X_train_new)
```

```python
# Dropping the variable 'sedan' and updating the model

X_train_2 = X_train_new.drop(["sedan"], axis = 1)

# Creating a model
lr_model(y_train,X_train_2)
VIF(X_train_2)
```

```python
# Dropping the variable 'hatchback' and updating the model

X_train_2 = X_train_2.drop(["hatchback"], axis = 1)

# Creating a model
lr_model(y_train,X_train_2)
VIF(X_train_2)
```

```python
# Dropping the variable 'curbweight' and updating the model

X_train_2 = X_train_2.drop(["curbweight"], axis = 1)

# Creating a model
lr_model(y_train,X_train_2)
VIF(X_train_2)
```

```python
# Dropping the variable 'wagon' and updating the model

X_train_2 = X_train_2.drop(["wagon"], axis = 1)

# Creating a model
lr_model(y_train,X_train_2)
VIF(X_train_2)
```

```python
# Dropping the variable 'porsche' and updating the model

X_train_2 = X_train_2.drop(["porsche"], axis = 1)

# Creating a model
lr_model(y_train,X_train_2)
VIF(X_train_2)
```

### <font color='red'>So now all the variables have p-value less than 0.05(Significance level) and vif less than 5.</font>
### <font color='red'>So this is our final linear regression model.</font>

## Step 7: Residual Analysis of the train data

So, now to check if the error terms are also normally distributed (which is infact, one of the major assumptions of linear regression), let us plot the histogram of the error terms and see what it looks like.

```python
X_train_lm = sm.add_constant(X_train_2)
lm = sm.OLS(y_train,X_train_lm).fit()   # Running the linear model
y_train_price = lm.predict(X_train_lm)
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
```

## Step 8: Making Predictions Using the Final Model

Now that we have fitted the model and checked the normality of error terms, 
it's time to go ahead and make predictions using the final model.

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['wheelbase', 'carlength', 'carwidth', 'curbweight','enginesize', 'boreratio',
            'horsepower', 'citympg', 'highwaympg', 'price']

cp_test[num_vars] = scaler.transform(cp_test[num_vars])
```

#### Dividing into X_test and y_test

```python
y_test = cp_test.pop('price')
X_test = cp_test
```

```python
# Now let's use our model to make predictions.

# Creating X_test_new dataframe by dropping variables from X_test
X_test_new = X_test[X_train_2.columns]

# Adding a constant variable 
X_test_new_lm = sm.add_constant(X_test_new)

# Making predictions
y_pred = lm.predict(X_test_new_lm)
```

## Step 9: Model Evaluation

Let's now plot the graph for actual versus predicted values.

```python
# Plotting y_test and y_pred to understand the spread.
fig = plt.figure()
plt.scatter(y_test,y_pred)
fig.suptitle('y_test vs y_pred', fontsize=20)              # Plot heading 
plt.xlabel('y_test', fontsize=18)                          # X-label
plt.ylabel('y_pred', fontsize=16)                          # Y-label
```

```python
# Lets calculate the R-squared score on the test set.

from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

<font color='red'>**So on test set we have r2_scrore of `0.88` which is good.**</font>
